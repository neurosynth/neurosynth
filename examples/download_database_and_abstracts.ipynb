# Downloading the Neurosynth database and abstracts from the internet

In this example we will use `neurosynth.dataset` to download the Neurosynth database (`download`), as well as the abstracts for all of the articles in the database (`download_abstracts`). Make sure to use your own email.

```python
import sys
from os.path import abspath, join
sys.path.append(abspath('..'))
import neurosynth
```

```python
data_dir = '/Users/tsalo/Desktop/ns-dataset/'
email = 'tsalo006@fiu.edu'
```

```python
neurosynth.base.dataset.download(data_dir, unpack=True)
```

```python
dset = neurosynth.base.dataset.Dataset(join(data_dir, 'database.txt'))
```

```python
abstracts_df = neurosynth.base.dataset.download_abstracts(dset, path=data_dir, email=email)
```
